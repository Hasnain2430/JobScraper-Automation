## Question 1

After successfully completing your degree, hopefully within eight semesters xD, you're embarking on the job search phase. Traditional job hunting can be a bit boring, so why not consider automation as a solution? Luckily, your PAI class introduced you to Selenium, an automation tool.

You've thinking of developing a simple script to navigate rozee.pk and indeed websites, searching for AI/ML Engineer positions. This script extracts crucial information from each job listing, such as job titles, company names, job locations, gender preferences, age requirements, minimum education levels, and experience details (if available). All this data should be neatly organized into a Pandas dataframe, including links to the job listings. Perform analysis on dataframe to find a job which best suits you.


```python
import pandas as pd
import time
import numpy as np

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import InvalidSelectorException

from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
import nltk
```

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
url = "https://www.rozee.pk/"
```

```python
driver.get(url)
search_bar = driver.find_element(By.XPATH,'/html/body/div[2]/div[3]/div[2]/form/div[1]/input')
search_bar.send_keys('AI/ML Engineer',Keys.RETURN)
time.sleep(5)
```

```python
job_title = []
company_name = []
job_locations = []
gender = []
min_edu = []
degree_lvl = []
exp = []
link = []

job_grid = driver.find_elements(By.CLASS_NAME,'job ')

for job in job_grid:
    try:
        jobs = job.find_element(By.CLASS_NAME,'s-18')
        job_link = jobs.find_element(By.TAG_NAME, "a")
        job_link = job_link.get_attribute("href")
        link.append(job_link)
        job_title.append(jobs.text)
        jobs.click()
        time.sleep(5)
        subwindow = driver.find_elements(By.CLASS_NAME,'col-12')
        for i in subwindow:
            info = i.text.splitlines()
            info = info[1].splitlines()
            info = info[0]
            info = info.split(',')[0].strip()
        company_name.append(info)
        subwindow = driver.find_element(By.XPATH,'/html/body/div[3]/div/div[2]/div[3]/div[1]/div/div[2]/div[3]/div')
        subwindow2 = subwindow.find_elements(By.CLASS_NAME,'row')

        count = 0
        for i in subwindow2:
            data = i.text.splitlines()
            heading = data[0]
            if (heading=='Job Location:'):
                data = data[1]
                job_locations.append(data)
            elif (heading=='Gender'):
                data = data[1]
                gender.append(data)
            elif (heading=='Minimum Education'):
                data = data[1]
                min_edu.append(data)
            elif (heading=='Degree Title'):
                data = data[1]
                degree_lvl.append(data)
            elif (heading=='Minimum Experience'):
                data = data[1]
                exp.append(data)
            count = count + 1
        if (len(gender)!=len(degree_lvl)):
            degree_lvl.append("-")
        if (len(gender)!=len(exp)):
            exp.append("-")
        # print("Job Title: ", jobs.text)
    except NoSuchElementException:
        pass
```

```python
data = {'Job Title': job_title, 'Company Name': company_name,'Job Location(s)':job_locations,'Gender Prefrence':gender,'Minimum Education':min_edu,'Degree Level':degree_lvl,'Experience ':exp,'Link':link}
df = pd.DataFrame(data)
```

```python
df.to_csv("Rozee_Data.csv")
```

```python
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
url = 'https://pk.indeed.com/'
driver.get(url)
```

```python
search_bar = driver.find_element(By.XPATH,'/html/body/div/div[1]/div/span/div[4]/div[1]/div/div/div/div/form/div/div[1]/div/div[1]/div/span/input')
search_bar.send_keys('AI/ML Engineer',Keys.RETURN)
time.sleep(5)
```

```python
job_title = []
company_name = []
job_locations = []
salary = []
job_type = []
link = []

num_pages_to_scrape = 3

for page in range(num_pages_to_scrape):

    job_grid = driver.find_elements(By.CLASS_NAME,'resultContent')
    for jobs in job_grid:
        job_link = jobs.find_element(By.TAG_NAME, "a")
        job_link = job_link.get_attribute("href")
        link.append(job_link)
        job = jobs.text.splitlines()
        title = job[0]
        job_title.append(title)
        company = job[1]
        company_name.append(company)
        loc = job[2]
        job_locations.append(loc)
        if (len(job)>3):
            sal = job[3]
            if 'a month' in sal:
                salary.append(sal)
                type = job[4]
                job_type.append(type)
            else:
                job_type.append(sal)
                salary.append('-')
        else:
            salary.append('-')
            job_type.append('-')

    try:
        driver.execute_script('window.scrollTo(0,5000) ' )
        time.sleep(5)
        button=driver.find_element(By.XPATH,"/html/body/main/div/div[1]/div/div[5]/div[1]/nav/ul/li[4]/a")
        button.click()
        time.sleep(5)
    except NoSuchElementException:
        break
```

```python
data = {'Job Title': job_title, 'Company Name': company_name,'Job Location(s)':job_locations,'Job Type':job_type,'Salary':salary,'Link':link}
df2 = pd.DataFrame(data)
df2.to_csv("Indeed_Data.csv")
```
